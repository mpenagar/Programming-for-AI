## Multimedia and Rich Output

Markdown supports $\LaTeX$ math:

* Inline equations, such as $E = m c^2$ or $\sum_{i=1}^{n} i = \frac{n(n + 1)}{2}$

* Standalone equations: $$E = mc^2 \;\;,\;\; \sum_{i=1}^{n} i = \frac{n(n + 1)}{2}$$

* Multi line aligned equations:
$$\begin{align*}
\sum_{i=1}^{n} (ax_i + b) &= \sum_{i=1}^{n} ax_i + \sum_{i=1}^{n} b \\
&= a \sum_{i=1}^{n} x_i + nb
\end{align*}$$

* Equations with cases:
$$Fib_n = 
\begin{cases} 
0 & \text{if } n = 0 \\
1 & \text{if } n = 1 \\
Fib_{n-1} + Fib_{n-2} & \text{if } n > 1
\end{cases}
$$

## Jupyter/iPython Commands

* Help: `?keyword`; for instance, `?np.linspace`, `?pd.read*`
* Shell commands are interpreted by prefixing them with `!`: `!ls`, `!pwd`, etc.
* Tab completion:
    * `<string>Tab` → compatible variable names,
    * `<variable>.Tab` → list of methods that can be applied to an existing variable
* History of commands: `history`
* `%reset` deletes the whole namespace
* `%xdel <variable>` tries to delete all references to a given variable
* `%save my_file.py ih[n]` saves the contents of cell n in <tt>my_file.py</tt>
* `%run my_file.py` runs the code stored in <tt>my_file.py</tt>
* `%time <statement>` executes and prints the running time of a statement
(typically a function call)
* `%timeit <statement>` executes it several times and shows the mean/std time
* `%prun <statement>` gives detailed profiler information
* `%who` shows the names of the variables defined in the runtime environment
* `%who <datatype>` filters the names by data type
* `%who_ls` returns a list of variable names
* `%whos` shows complete info about the defined variables
