## Lab 1 - Python

**1)** Write a function `mystr(number)` that takes a natural number and returns it's text representation in decimal format. Do not use the built-in `str` function.

**2)** Write a function `greatest_digits(iterable)` that, given an iterable that contains natural numbers, returns the number with the greatest sum of it's digits (in decimal format). For instance, if the input list is `[32, 19, 10012, 587, 2001]`, the function should return `587`.

**3)** Write a function `greatest_digits(iterable)` that, given an iterable that contains natural numbers, returns a list where integers are sorted according to the sum of their digits (in decimal format). For instance, if the input list is `[32, 19, 10012, 587, 2001]`, the function should return `[2001, 10012, 32, 19, 587]`.

**4)** Write a generator function `real_range(start, stop, step=1.0)` that returns the sequence of real values $\{start, \;start+step, \;start+2 step, \;\dots, \;a+k \cdot step\}$ where $k$ is the greatest integer such that $start+k \cdot step \,\le\, stop$

**5)** Write a function `get_max(func, a, b, step=0.001)` that searches the minimum of another function $func(x)$ for $x \in [a,b]$ and the given search step (i.e., the minimum of $func(x)$ for $x \in \{a, \;a+step, \;a+2 step, \;\dots, \;a+k \cdot step\}$

**6)** Write a function `definite_integral(func, a, b, step=0.001)` that approximates the definite integral of the given function in the interval $[a,b]$ and the given $step$ by:
$$\int_a^b{func(x) \; dx} \approx \sum_{x \in \{a, \;a+step, \;a+2 step, \;\dots, \;a+k \cdot step\}}{func(x) \cdot step}$$

**7)** Write a generator function `my_range(start, stop=None, step=1)` that emulates the built-in `range()` function. If a single positional argument is given, it will be used as `stop` and `start=0`.

**8)** Write a generator function `my_enumerate(iterable, start=0)` that emulates the built-in `enumerate()` function.

**9)** Write a generator function `naturals(start=1)` that returns the sequence of all natural numbers starting from `start`.

**10)** Write a function `next_prime(previous_list)` that receives a list of ascending prime numbers `[2, 3, 5, 7, 11, ...]` and returns the next prime number.

**11)** Write a generator function `primes()` that returns the infinite sequence of prime numbers.

**12)** Create a function `most_frequent_chars(filename, encoding="utf8", n=10)` that returns a list with the $n$ most frequent characters of a text file, ordered from most to less frequent.

**13)** Try to verify (with code) that the functions `random.random()` and `random.randint(a,b)` truly return values following uniform distributions.
