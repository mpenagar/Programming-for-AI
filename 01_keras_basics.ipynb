# Neural Network abstraction

## Connectionist model built from complex interactions of simple units

- **Basic Model**: Multi-layer perceptron (MLP). Dense connections of units across multiple layers:
    - **Neurons**: Basic units that receive input, process it, and pass on the output.
    - **Layers**: 
      - **Input Layer**: Where data enters.
      - **Hidden Layers**: Intermediate layers where computation happens.
      - **Output Layer**: Final layer that outputs results.
    - **Activation Function**: Determines the output of neurons; common functions include Sigmoid, Tanh, and ReLU.

## Advantages:
 - **Hardware-friendly**: Regular structures, well suited for optimization (tensor-based computations, parallelization).
 - **Scalability**: You can train **huge models** on **huge datasets** (trillions of parameters over TBs of data).
 - **They can manage unstructured data**: Usually you don't need **feature engineering**, they can extract them automatically from the raw data (images, audio, text...).

## Disadvantages:
 - **Data-hungry**: Their benefits are usually seen only on **extremely large** datasets.
 - **Black boxes**: Predictions cannot be explained or interpreted. Potential unexpected behaviors. No guarantees.
 - **Poor handling of discontinuities**: Smoothness is assumed in the feature space and output. Difficulties with discrete problems (rule-based classificiations, counting, etc.)

## 1. First Neural-Network example in Keras:

### The "hello world" of ML: Classifying the Iris Dataset

 - Let's see a basic example of how NNs are also suitable for feature-based classification.

```python
#Configuration of Keras backend (JAX) using only CPU
import os
os.environ["KERAS_BACKEND"] = "jax"
os.environ["JAX_PLATFORMS"] = "cpu"
```

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import keras
```

```python
# Loading dataset
iris = load_iris()
X = iris.data
y = iris.target
print(iris.DESCR)
```

```python
# Splitting data to train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Scaling features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create the model
model1 = keras.models.Sequential([
    keras.layers.Input(shape=(4,)),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(6, activation='relu'),
    keras.layers.Dense(3, activation='softmax')
])

# Compile the model
model1.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
model1.summary()
```

```python
#Fit the model and see the training behavior
history1 = model1.fit(X_train, y_train, epochs=50, batch_size=5, validation_split=0.2)
```

```python
#Learning curves
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
plt.show()
```

```python
# Evaluate the model
y_pred_probs = model1.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

print("Accuracy on test set:", accuracy_score(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

## 2: Avoiding feature engineering: A CNN example on the MNIST dataset

- **Type**: Handwritten Digit Dataset
- **Contents**:
  - **Training Set**: 60,000 images
  - **Test Set**: 10,000 images
- **Image Details**:
  - **Format**: Grayscale
  - **Size**: 28x28 pixels
- **Classes**: Digits 0 through 9
- **Applications**:
  - Handwritten digit recognition
  - Basic machine learning and deep learning experiments


### Data exploration

```python
# Load the data and split it between train and test sets
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
# CNNs require 4 input dimensions (samples, width, height, channels)
# Thus, make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")
```

### Data visualization: Average image for each class.

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
# Loop over the 2x5 grid and plot something on each subplot
d = 0
for i in range(2):
    for j in range(5):
        samples_digit = x_train[y_train==d]
        axes[i, j].imshow(np.mean(samples_digit, axis=0), cmap='Greys')
        axes[i, j].set_title(f'Average Digit {d}')
        d += 1
# Adjust layout for better spacing between plots
plt.tight_layout()
# Display the plots
plt.show()
```

### Model definition

```python
# Model parameters
num_classes = len(set(y_train))
input_shape = (28, 28, 1)

model = keras.Sequential(
    [
        keras.layers.Input(shape=input_shape),
        keras.layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        keras.layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        keras.layers.MaxPooling2D(pool_size=(2, 2)),
        keras.layers.Conv2D(128, kernel_size=(3, 3), activation="relu"),
        keras.layers.Conv2D(128, kernel_size=(3, 3), activation="relu"),
        keras.layers.GlobalAveragePooling2D(),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(num_classes, activation="softmax"),
    ]
)
model.summary()
```

### Model Training and Evaluation

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
batch_size = 128
epochs = 4

callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=2),
]

history = model.fit(
    x_train,
    y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=0.15,
    callbacks=callbacks,
)
```

```python
score = model.evaluate(x_test, y_test)
print(f'Test accuracy: {score[1]} - Test loss: {score[0]}')
```
