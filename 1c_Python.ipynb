## Numbers - `int` & `float`

* **Integers**: `int`
    * Can take arbitrarily large values: the internal representation grows
on demand (i.e., the integer size is not fixed)
    * Arithmetic operators: `+`, `-`, `*`, `//`, `%`, `**`
    * Extended assignments: `+=`, `-=`, `*=`, `//=`, `%=`, `**=`
    * Integer casting: `int(float | str)`
    * Builtin functions: `abs(a)`, `pow(a,b,mod=None)`, divmod(a,b), hex(a), oct(a)
* **Floating-point numbers**: `float`
    * Fixed size of 64 bits (i.e. their representation range & precision is limited)
    * Arithmetic operators: `+`, `-`, `*`, `/`, `**`
    * Extended assignments: `+=`, `-=`, `*=`, `/=`, `**=`
    * Float casting: `float(int | str)`
    * Built-in functions: `abs(x)`, `pow(x,y)`, round(x)

## Booleans - `bool`

* `True` and `False` values
* Result of relational operators: `>`, `>=`, `<`, `<=`, `==`, `!=`
* Operands of boolean operators: `and`, `or` and `not`
* Multiple relations: `a<b<c` is the same as `a<b and b<c`
* Boolean casting: `bool(something)`
    * `None` &rarr; `bool(None) == False`
    * Numbers:
        * `bool(<zero value>) == False`
        * `bool(<non-zero value>) == True`
    * Sequences or collections
        * `bool(<empty>) == False`
        * `bool(<non-empty>) == True`
    * In conditional instructions (`if`, `elif`, `while`), the `bool()` function
is implicitly applied.
        * Any object can be used as a condition

````python
for x in [0,1,0.0,1.0,1e-30,"Hi","",[1,2,3],[],range(10),range(0)] :
    if x :
        print(x, "is True:", bool(x))
    else :
        print(x, "is False:", bool(x))
````

## Strings - `str`

* A sequence of characters (enclosed in quotes)
```python
s = "Hello, world!"
t = 'Bye, bye, world!'
```
* Triple-quoted strings (docstrings) may span several lines:
```python

```
* Special characters: `'\t'`, `'\n'`, ...
* Strings are **immutable**: once created, <u>they cannot be modified</u>
* `len(s)`: number of characters
* String operators:
    * `+` : concatenation `"Hello" + ", " + "world!"`
    * `*` : repeated concatenation `"Hello world!" + 10`
    * `in` `not in` : check the presence of some substring `"wor" in "Hello, world!"`

### Strings - Indexing & Slice Notation

* A string can be indexed to get a single character out of it: `s[i]`
    * Positive indexes in the range `[0 , len(s)-1]`
    * Negative indexes in the range `[-len(s) , -1]` ($i<0 \to i\equiv len(s)+i$)
    * The result is a new string of length 1
* **Slice Notation**: extracting a specific range of chars from a string
    * `s[start:stop:step]`
        * `start` is the inclusive beginning index
        * `stop` is the exclusive ending index
        * `step` defines the increment between elements
    * No `step` &rarr; `step=1`
    * No `start` &rarr; `start=0` (_from the beginning_)
    * No `stop` &rarr; `stop=len(s)` (_to the end_)


````python
s = "Hello world!"
print(s[4])
print(s[-4])
print(s[2:10])
print(s[:4])
print(s[-4:])
print(s[::2])
print(s[::-1])
````

## Lists - `list`

* A **mutable** sequence of values of any type
```python
z = [3, 1.98, -15, "Peter", ['a', "b", 'c'], 1]
```
* `list(it)`: creates a list with the items of an **iterable** object `it`
    * Strings are iterable &rarr; `list("aeiou")` &rarr; `['a','e','i','o','u']`
* Empty list: `[]` or `list()`
* `len(z)`: number of items
* Lists are indexable, just like strings, and slice compatible.
* List operators:
    * `z1 + z2` : concatenation 
    * `z * n` : repeated concatenation
    * `x in z` (and `x not in z`): `True` if any of the items of `z` is equal to `x` 

### Lists - methods

* **Simple modification**: `z[<index>] = <new value>`
* **Search & information**
    * Index of a value: `z.index(<value>)`
    * Number of occurrences of a value: `z.count(<value>)`
* **Adding/Removing items**
    * Insert value at given position: `z.insert(<index>,<value>)`
    * Add to the end: `z.append(<value>)`
    * Extend (appending the items of an *iterable*): `z.extend(<iterable>)`
    * Remove a value: `z.remove(<value>)`
    * Delete item at given position: `del z[<index>]` (slice compatible: `del z[:5]`)
    * Delete and get: `z.pop([<index>])` (default `index=-1`)
    * Empty the list: `z.clear()`
* **Reordering**
    * Reverse items: `z.reverse()`
    * Sort items: `z.sort(key=None, reverse=False)`
* **Copying (shallow)**: `z.copy()` or `z[:]` 

### Lists - built-in functions

Some built-in functions can be used with or return lists.

* **Ordering**
    * `sorted(iterable, key=None, reverse=False)` : Returns a new ordered list

## Tuples - `tuple`

* An **immutable** sequence of values of any type
```python
z = (3, 1.98, -15, "Peter", ['a', "b", 'c'], 1)
```
* `tuple(it)`: creates a tuple with the items of an **iterable** object `it`
* Empty tuple: `()` or `tuple()`
* Single item tuple: `(<value>,)`
* `len(t)`: number of items
* Tuples are indexable, just like lists, and slice compatible.
* Tuple operators:
    * `t1 + t2` : concatenation 
    * `t * n` : repeated concatenation
    * `x in t` (and `x not in t`): `True` if any of the items of `t` is equal to `x` 

### Tuples - Sequence Unpacking

* Python allows multiple assignment:
    * `<var1>,<var2>,... = <value1>,<value2>,...`
* The left side of the assignment can be a tuple/list of variables
* The right side of the assignment can be any iterable
    * The iterable is *unpacked* into the variables
    * The number of items of the iterable must match the number of variables
* **Nested Sequence Unpacking**:
    * `<var1>,(<var2>,<var3>,...),... = <value1>,(<value2>,<value3>,...),...`

````python
a,b,c = 1,2,3
print(f'     a,b,c = 1,2,3 --> {a=} {b=} {c=}')
a,b,c = (1,2,3)
print(f'   a,b,c = (1,2,3) --> {a=} {b=} {c=}')
a,b,c = [1,2,3]
print(f'   a,b,c = [1,2,3] --> {a=} {b=} {c=}')
a,b,c = range(1,4)
print(f'a,b,c = range(1,4) --> {a=} {b=} {c=}')
a,b,c = "123"
print(f'     a,b,c = "123" --> {a=} {b=} {c=}')
(a,b,c) = 1,2,3
print(f'   (a,b,c) = 1,2,3 --> {a=} {b=} {c=}')
[a,b,c] = 1,2,3
print(f'   [a,b,c] = 1,2,3 --> {a=} {b=} {c=}')
````

Unpacking size mismatch errors:

````python
# ValueError: too many values to unpack (expected 3)
#a,b,c = range(4)

# ValueError: not enough values to unpack (expected 3, got 2)
#a,b,c = range(2)
````

Nested Sequence Unpacking:

````python
a,(b,c) = 1,(2,3)
print(f'a,(b,c) = 1,(2,3) --> {a=} {b=} {c=}')
a,(b,c) = 1,"23"
print(f'a,(b,c) = 1,"23" --> {a=} {b=} {c=}')
a,(b,c) = 1,range(2)
print(f'a,(b,c) = 1,(2,3) --> {a=} {b=} {c=}')
````

### Tuples - Immutability

* Tuples (like lists) do not contain objects/values, but **references** to them
* A tuple is an **immutable sequence of references**
   * It cannot change which objects it points to
   * The state (*value*) of those objects (if they are mutable) can change.
      * The state of the tuple can change

````python
my_tuple = (42, ["apple", "banana"])
print(f"{my_tuple = }")
# ❌ ILLEGAL: Trying to change the TUPLE
#my_tuple[1] = ["cherry"] 
# ✅ LEGAL: Changing the a mutable object of the tuple
my_tuple[1].append("cherry") 
print(f"{my_tuple = }")
````

### More Built-in Functions - `reversed`

<div class="nota-flotante">
    
```python
reversed(object)
```
</div>

* Creates an iterable representing the reverse item sequence
* Not all objects can be reverted
    * Those internally defining the *reversed method*
    * Those following the **sequence Protocol**:
        * `len(object)` & `object[i]` for $\forall i \in [0,len(object)-1]$
        * `str`, `list` & `tuple`
* **Lazy Iterable**: items are created <u>on demand</u> (when iterated)
    * `a_list[::-1]` vs `reversed(a_list)` &rarr; does not create a copy of the entire list

````python
z = list(range(20))
for i in reversed(z):
    print(i, end=" ")
````

### More Built-in Functions - `enumerate`

<div class="nota-flotante">
    
```python
enumerate(iterable, start=0)
```
</div>

* Creates an iterable sequence representing an enumeration
    * Yields pairs containing a count and a value yielded by the iterable.
* It is exhausted when the iterable is exhausted.
* **Lazy Iterable**: items are created on demand (when iterated)

````python
z = ["blue", "green", "white", "black"]
print(enumerate(z))
````

````python
for pair in enumerate(z):
    print(f'{pair = }')
````

````python
for i,x in enumerate(z):
    print(f'{i} --> {x}')
````

````python
print(list(enumerate(z)))
````

**enumerate** is useful when we want both the index and the value of the items in an *iterable*.

Let's suppose we want to get the index of the greatest item of an iterable: 

````python
z = [3,4,1,6,3,5]

i_max = 0
x_max = z[0]
for i,x in enumerate(z):
    if x > x_max :
        i_max = i
        x_max = x

print(f'The greatest item is at index={i_max}')
````

### More Built-in Functions -  `zip`

<div class="nota-flotante">
    
```python
zip(*iterables, strict=False)
```
</div>

* Creates an iterable sequence containing pairs, trios, quartets...
    * Yields n-length tuples (n = number of iterables).
    * The i-th element in every tuple comes from the i-th iterable
* It is exhausted when the shortest iterable is exhausted.
    * If `strict=False` ensures same length (or `ValueError`)
* **Lazy Iterable**: items are created on demand (when iterated)

````python
z1 = ["blue", "green", "white",  "black"]
z2 = ["azul", "verde", "blanco", "negro"]
print(zip(z1,z2))
````

````python
for pair in zip(z1,z2):
    print(f'{pair = }')
````

````python
for en,es in zip(z1,z2):
    print(f'{en} --> {es}')
````

````python
print(list(zip(z1,z2)))
````

**zip** is useful when we want to simultaneously iterate through two or more *iterables*.

Let's suppose we have two lists, with the names and phone numbers of our friends and we want to get the phone number of one of them: 

````python
names = ["Ander", "Jone", "Sara", "Jon"]
numbers = [645345274, 655856394, 664374566, 639998273]
search_name = "Sara"

for name,number in zip(names,numbers):
    if name == search_name :
        search_number = number
        break

print(f'The phone number of {search_name} is {search_number}')
````

## Dictionaries - `dict`

* A **mutable** mapping with `<key>` &rarr; `<value>` entries
    * Keys must have <u>immutable value/state</u>
    * Values can be anything 
```python
d = {"Peter":45 , 34:"Alice" , "Lyn":[1,2,"hi"] , (1,2):"Joe"}
```
* `dict(it)`: creates a dict with the `<key>,<value>` items of an iterable
* `dict(d)`: creates a dict with `<key> in d` &rarr; `d[<key>]` entries
* `dict(<key1>=<value1>,...)`: creates a dict with `"<key>"` &rarr; `<value>` entries
* `dict.fromkeys(it, value=None)`: creates a dict with keys from `it` and all value=`value`
* Empty dictionary: `{}` or `dict()` 
* `len(d)`: number of `<key>,<value>` entries
* Dictionaries are indexable (with `<key>`s) but <u>not slice compatible</u>.
    * `d["Peter"]` or `d.get("Peter")` or `d.get("Peter", default=<something>)`
* Dictionaries are **iterable**: <u>sequence of `<key>`s</u>
* Dictionary operators:
    * `d1 | d2` : merge
    * `d1 |= d2` or `d1.update(d2)` :  update
    * `k in d` (and `x not in d`): `True` if dictionary `d` contains key `k` 

### Dictionaries - methods

* **Simple modification**: `z[<key>] = <new value>`
* **Information**
    * Sequence of `<key>`: `d.keys()` (equivalent to `d`)
    * Sequence of `<value>`: `d.values()`
    * Sequence of `<key>,<value>`: `d.items()`
    * `d.setdefault(key, default=None)` : Set `d[key]=default` if `key not in d` and return `d[key]`
* **Adding/Removing items**
    * Update with another dict-like: `d.update(d2)`
    * Delete an entry by key: `del d[<key>]`
    * Delete an entry by key and get (the removed value): `d.pop(<key>)`
    * Delete some entry and get (the removed entry): `d.popitem()`
    * Empty the dict: `d.clear()`
* **Copying (shallow)**: `d.copy()`

## Sets - `set`

A generator that returnn only the first `n` (default=10) values of any iterable:

````python
def first(it, n=10):
    """yields the first n (default 10) elements from an iterable"""
    for x in it:            
        yield x
        n -= 1
        if n <= 0 :
            break
````

````python
s = "this is a text with many different characters in it"
print('>', *first(s), '<', sep="")
print('>', *first(s,15), '<', sep="")
print('>', *first(s,100000), '<', sep="")
# fails, returns first value
print('>', *first(s,0), '<', sep="")
````

````python
def first(it, n=10):
    """yields the first n (default 10) elements from an iterable"""
    for _,x in zip(range(n),it):
        yield x
````

A tricky matrix determinant calculus: $$|A| = \sum_{j=0}^{n-1} (-1)^{i+j} \cdot a_{ij} \cdot |A_{ij}| \,\xlongequal{\text{i=0}}\, \sum_{j=0}^{n-1} (-1)^{j} \cdot a_{0j} \cdot |A_{0j}|$$

````python
def plusminus(): # sequence of  1, -1, 1, -1, ...
    pm = 1
    while True :
        yield pm
        pm = -pm

def row_0_minors(A): # Sequence of minors of A, developed over first row
    n = len(A)
    A = A[1:] # "remove" first row
    for i in range(n):
        yield [[r[j] for j in range(n) if j!=i] for r in A]

def det(A):
    if len(A) ==1 :
        return A [0][0]
    else :
        factors = zip(plusminus(), A[0], row_0_minors(A))
        return sum(sign*a*det(minor) for sign,a,minor in factors )
````

* `uniform(a, b)`: A real number uniformly distributed in the interval $[a,b)$
* `normalvariate(mu=0.0, sigma=1.0)`: A real number from a Normal distribution $\mathcal{N}(\mu=mu,\,\sigma=sigma)$

* `with`: a **context manager** that simplifies resource management
    * Handles setup and cleanup tasks
    * Ensures that allocated resources are properly released
```python
with <resource allocation> as <variable>:
    body # use here the resource (<variable>)
# the resource has been properly released
```
* `with` statement for file handling:
```python
with open(path, mode="r", encoding=None) as file:
    body # use the file
# the file has been properly closed
```

````python
alphabet="abcdefghijklmnopqrstuvwxyz "
s = set()
with open("bible.txt", mode="r", encoding="utf8") as f:
    for line in f:
        # `\n` will be removed because it is not in the alphabet
        #line = line.rstrip()
        line = "".join((c if c in alphabet else " ") for c in line.lower())
        s.update(line.split())
print(f"There are {len(s)} different words in the bible")
````

## Exceptions

* Events that <u>disrupt</u> the normal flow of a program when an error occurs.
    * Instead of crashing, an exception object is **triggered/raised**
    * Errors can be **caught** and managed
```python
try :
    body # execptions could be raised here 
except <error_type_1> :
    body # executed if exception of type <error_type_1> caught
except <error_type_> as <variable> :
    body # executed if exception of type <error_type_2> caught
except (<error_type_3>,<error_type_4>) :
    body # executed if exception of type <error_type_3> or <error_type_4> caught
except: # equivalent to except BaseException :
    body # manages any kind of exception.
else:
    body # executed if there is no exception
finally:
    body # executed ALWAYS
```

## Classes 

* Blueprints for creating new object types.
* The foundation of Object-Oriented Programming (OOP)
* An object bundles
    * Data &rarr; **Attributes**
    * Functionality &rarr; **Methods**
* Class definition &rarr; new object-type creation
* Class *execution* &rarr; new object invocation
* All classes extend another class (default: `object`)


### Classes - definition

```python
class <class_name>(<extended_class_name>):
    """ <Docstring> """
        
    def methodA(self, <args>):
        body
        
    def methodB(self, <args>):
        body
```
* Class methods have typically a first special argumen `self`
    * A reference to the object that invoked the method.
    * Arbitrary name, but <tt>self</tt> is standard
    * Method arguments (plus `self`) are passed to the method
        * List `z`: `z.append(x)` &rarr; `list.append(z,x)`
        * Dictionary `d`: `d.get(k,default=0)` &rarr; `dict.get(d,k,default=0)`

### Classes - `__init__` method

```python
class <class_name>(<extended_class_name>):
    """ <Docstring> """
    
    def __init__(self, a, b, c):
        initialization body
```
* Special method `__init__()` is executed when we create a new object
    * *Constructor* method
    * Instantiation arguments (plus `self`) are passed to `__init__()`
    * `MyClass(a_value, b_value, c_value)`:
        1. Create object `o` of `type(MyClass)`
        2. Execute `MyClass.__init__(o, a_value, b_value, c_value)`
        3. Return initialized object `o`
* We can dinamically create new attributes:
    * `self.<attribute_name> = <value>`
    * Object attributes are typically created inside the constructor

````python
class Robot:
    """A simple Robot class"""

    # The "Constructor" - runs when you create a new robot
    def __init__(self, name, model_year):
        self.name = name            # An instance attribute
        self.model_year = model_year # An instance attribute

    # A custom method (behavior)
    def say_hello(self):
        print(f"Beep boop! I am {self.name}, built in {self.model_year}.")

# Creating an 'instance' (an actual object) from the blueprint
robot = Robot("WALL-E", 2805)

# Accessing data and behavior
robot.say_hello() 
````

### Classes - Special Methods

* Their names start and end with two underscores, `__`
* Many built-in functions delegate to/rely on such functions:
    * `len(obj)` &rarr; `obj.__len__()` &rarr; `<class of obj>.len(o)`
* Some examples:
    * `len()` &rarr; `__len__(self)`: Returns the "length" of the object
    * `str()` &rarr; `__str__(self)`: Returns a user-friendly string representation

````python
robot = Robot("WALL-E", 2805)
print(robot)
````

````python
class Robot:
    """A simple Robot class"""

    # The "Constructor" - runs when you create a new robot
    def __init__(self, name, model_year):
        self.name = name            # An instance attribute
        self.model_year = model_year # An instance attribute

    # A custom method (behavior)
    def say_hello(self):
        print(f"Beep boop! I am {self.name}, built in {self.model_year}.")

    # implementing str()
    def __str__(self):
        return f"Robot(name='{self.name}', model_year={self.model_year})"        

# Creating an 'instance' (an actual object) from the blueprint
robot = Robot("WALL-E", 2805)
print(robot)
````

### Classes - Special Methods

* Operators, just like built-in method, delegate to/rely on such functions
    * `obj1 + obj2` &rarr; `obj1.__add__(obj2)` &rarr; `<class of obj>.__add__(obj1, obj2)`
* Arithmetic Operators examples:
    * `+` &rarr; `__add__(self, other)`: Addition
    * `-` &rarr; `__sub__(self, other)`: Subtraction
    * `*` &rarr; `__mull__(self, other)`: Multiplication
* Comparison Operators examples:
    * `==` &rarr; `__eq__(self, other)`: Equality
    * `!=` &rarr; `__ne__(self, other)`: Not Equal
    * `<` &rarr; `__lt__(self, other)`: Less Than
    * `<=` &rarr; `__le__(self, other)`: Less or Equal
* Indexing Operators examples:
    * `obj[key]` &rarr; `__getitem__(self, key)`: Get
    * `obj[key] = value` &rarr; `__setitem__(self, key, value)`: Set
    * `del obj[key]` &rarr; `__detitem__(self, key)`: Delete

````python
class Point2D():
    """A 2 dimensional Point"""

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point2D(x={self.x}, y={self.y})"

    def __add__(self, other):
        if type(other) == Point2D:
            return Point2D(self.x+other.x, self.y+other.y)
        elif type(other) in (int, float) :
            return Point2D(self.x+other, self.y+other)
        else :
            raise TypeError(f"Cannot add a Point2D and {type(other)}")
````

````python
p1 = Point2D(1,2)
p2 = Point2D(10,20)
print(p1, p2)
print(p1+p2)
print(p1+100)
# TypeError: Cannot add a Point2D and <class 'str'>
#print(p1+"100")    
````

### Classes - Reflected Special Methods

* We cannot add an `int` and a `Point2D`, because `int`s don't know about `Point2D` class, its `int.__add__(self, other)` method will fail:

````python
# unsupported operand type(s) for +: 'int' and 'Point2D'
#print(100+p1)
````

In those cases Python looks at the `Point2D` object for a reflected method:

`my_object.__radd__(<int>)`

````python
class Point2D():
    """A 2 dimensional Point"""

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point2D(x={self.x}, y={self.y})"

    def __add__(self, other):
        if type(other) == Point2D:
            return Point2D(self.x+other.x, self.y+other.y)
        elif type(other) in (int, float) :
            return Point2D(self.x+other, self.y+other)
        else :
            raise TypeError(f"Cannot add a Point2D and {type(other)}")

    def __radd__(self, other):
        # `self + other` is the same as `self.__add__(other)` 
        return self + other

p1 = Point2D(1,2)
print(15+p1)
````
