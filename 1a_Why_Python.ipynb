## Why Python?

* **Simple and clean**: Easy to learn (perfect for beginners), read and maintain
* **Interpreted**: quicker for development (at the expense of speed)
* **Dynamically typed**: Types checked at runtime, no need to declare variables explicitly
* **Versatile and resourceful**: the Python standard library includes classes and functions that allow to accomplish many different tasks.
* **Easy to extend**: Thousands of third-party packages extend Python’s capabilities almost to infinity
